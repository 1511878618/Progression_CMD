```python
import pandas as pd
from pathlib import Path
from config import *


# secondary preventive dir
data_dir = Path("data")

secondary_prevantative_combinations = [
    (i.parent.parent.stem, i.parent.stem)
    for i in list(data_dir.rglob("*disease.feather"))
]
secondary_prevantative_combinations
```

```python
df_list = []
for each in list(data_dir.rglob("*disease.feather")):
    df = pd.read_feather(each)[["eid", "event"]].rename(
        columns={"event": each.parent.parent.stem + "_to_" + each.parent.stem}
    )
    df_list.append(df)

from functools import reduce

df = reduce(lambda left, right: pd.merge(left, right, on="eid", how="outer"), df_list)

df
```

```python
regenie = df.rename(columns={"eid": "FID"})
regenie.insert(1, "IID", regenie["FID"])
regenie
```

```python
regenie_data_dir = outputDir / "regenie_files"
regenie_data_dir.mkdir(exist_ok=True, parents=True)
regenie.to_csv(
    regenie_data_dir / "regenie_data.regenie", sep="\t", index=False, na_rep="NA"
)
```
